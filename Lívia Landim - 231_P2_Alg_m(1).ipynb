###**1**.  [1,0 pontos] Em Python crie um programa que calcule e apresente o total de saldos cashback (float) de alguns clientes, mas não é informada a quantidade de clientes, então use como critério de parada (condição da estrutura de repetição, digitar zero no saldo cashback para sair).

```
saldo = float(input('Informe o saldo do cliente: (caso deseje encerrar, digite 0)'))
saldoTotal = 0

while saldo != 0 :
    if saldo > 0:
        saldoTotal = saldoTotal + saldo
        saldo = float(input('Informe o saldo do cliente: (caso deseje encerrar, digite 0)'))
    else :
        print('Não é possível cadastrar saldos negativos')

print(f'O total de saldos é: R${saldoTotal:.2f} ')
```

###**2**. [4,0 pontos] Em Python crie um programa que leia/armazene em um vetor 5 nomes de clientes, em outro vetor leia/armazene 5 valores/float de cashback e num terceiro vetor leia/armazene 5 valores que corresponam as compras que cada um realizou.
###No momento da compra, verifique se o saldo cashback do cliente é o suficiente para pagar a compra realizada, se sim subtraia do cashback o valor da compra realizada, caso não tenha, mostre a frase "Não há valor de cashback suficiente para pagamento desta compra."
###Ao final, mostre o vetor nome de clientes, o vetor saldo cashback e o vetor compras.


```
nome = []
cashback = []
compra = []

for i in range(5) :
    nome.append(input('Informe o nome do cliente: '))
    cashback.append(float(input('Informe o cashback desse cliente: ')))
    compra.append(float(input('Informe o valor da compra realizada: ')))

    print('')

    valorCompra = cashback[i] - compra[i]
    if valorCompra >= 0 :
        print(f'É possível realizar essa compra, {nome[i]}. O seu cashback ficará com um valor final de R$ {valorCompra:.2f}')
    else:
        print(f'Não há cashback suficiente para o pagamento dessa compra, {nome[i]}')

    print('-' * 50)

print('')

print(f'O vetor nome equivale a: {nome}')

print('-'*50)

print(f'O vetor cashback equivale a: {cashback}')

print('-'*50)

print(f'O vetor compra equivale a: {compra}')

print('-'*50)
```

###**3**. [2,0 pontos] Em Python crie um programa que leia/armazene em uma matriz 3x3 valores inteiros, calcule e apresente a soma dos elementos da diagonal principal.

```
matriz = []
somaElementos = 0
for lin in range(3):
    valor = []
    for col in range(3):
        valor.append(int(input('Informe um valor inteiro: ')))
    matriz.append(valor)

for lin in range(len(matriz)) :
    for col in range(len(matriz[0])) :
        if lin == col :
            somaElementos = somaElementos + matriz[lin][col]

print(f'A soma dos elementos da diagonal principal é igual a: {somaElementos}')
```
