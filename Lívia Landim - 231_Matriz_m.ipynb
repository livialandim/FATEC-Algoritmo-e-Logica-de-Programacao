### 2. Faça um programa que leia uma matriz 2 x 2, que representa o salário de 4 funcionários, calcule e mostre a soma total de todos os salários/elementos que será o montante pago pela empresa a esses funcionários.

### 4. Faça um programa que leia números inteiros w e p e os elementos de uma matriz A de números inteiros de dimensão w x p e conte o número de elementos que são iguais a zero. Observação w representa a quantidade de linhas e p as colunas


```
matriz = []
w = int(input('Informe um número inteiro para ser a quantidade de linhas: '))
p = int(input('Informe um número inteiro para ser a quantidade de colunas: '))

for lin in range(w):
    linha = []
    for col in range(p):
        linha.append(int(input('Informe um número inteiro: ')))
    matriz.append(linha)

numerosIguaisZero = 0

for lin in range(len(matriz)):
    for col in range(len(matriz[0])):
        if matriz[lin][col] == 0:
            numerosIguaisZero += 1

print(f'A quantidade de números iguais a zero é {numerosIguaisZero}')
```

### 5. Faça um programa que leia uma matriz 3 x 2, que representa preços de produtos, crie OUTRA matriz que armazene todos os preços com 7% de aumento.

```
matriz = []

for lin in range(3):
    linha = []
    for col in range(2):
        linha.append(float(input('Informe o preço do produto: ')))
    matriz.append(linha)

aumentoMatriz = []

for lin in range(len(matriz)):
    aumento = []
    for col in range(len(matriz[0])):
        calculo = matriz[lin][col] * 1.07
        aumento.append(calculo)
    aumentoMatriz.append(aumento)

print()

print('----- Matriz com preços originais -----')
for lin in range((len(matriz))):
    for col in range(len(matriz[0])):
        print(f'{matriz[lin][col]:.2f}', end='\t')
    print()
print()

print('----- Matriz com 7% de aumento -----')
for lin in range((len(aumentoMatriz))):
    for col in range(len(aumentoMatriz[0])):
        print(f'{aumentoMatriz[lin][col]:.2f}', end='\t')
    print()
```

### 6. Faça um programa que leia: 
###* um vetor com oito posições com os nomes das lojas; 
###* um outro vetor com quatro posições com os nomes dos produtos;
###* uma matriz (8 x 4) com os preços de todos os produtos em cada loja. 
###O programa deve mostrar todas as relações/print (nome da loja - nome do produto e preço), nas quais o preço não ultrapasse R$ 120,00.


```
# for lin até 8    lojas
#    for col até 4    produtos
matriz = []
nomeLoja = []
produto = []

for i in range(8):
    nomeLoja.append(input('Informe o nome da loja: '))

print('-'*50)
for i in range(4):
    produto.append(input('Informe o nome do produto: '))

print('-'*50)
for lin in range(len(nomeLoja)):
    preco = []
    for col in range(len(produto)):
        preco.append(float(input(f'Informe o preço do produto {produto[col]} da loja {nomeLoja[lin]}  ')))
    matriz.append(preco)

print('-'*50)
for lin in range(len(matriz)):
    for col in range(len(matriz[0])):
        if matriz[lin][col] <= 120:
            print(f'Loja: {nomeLoja[lin]} - Produto: {produto[col]} - Preço: {matriz[lin][col]}')
```

###7. Faça um programa que leia uma matriz 4 x 4. 
###* Apresente os valores/elementos e seus respectivos índices (linhas e colunas) que estão na diagonal principal.
###* Calcule e apresente a soma dos elementos acima da diagonal principal, que estão em negrito e com a cor de fundo amarela.

image.png


```
matriz = []

for lin in range(4):
    linha = []
    for col in range(4):
        linha.append(float(input('Informe um número: ')))
    matriz.append(linha)

for lin in range(len(matriz)):
    for col in range(len(matriz[0])):
        if lin == col:
            print(f'O elemento {matriz[lin][col]} faz parte da diagonal principal com os índices [{[lin]},{[col]}]')

somaElementos = 0

for lin in range(len(matriz)):
    for col in range(len(matriz[0])):
        if lin < col:
            somaElementos += matriz[lin][col]

print(f'A soma dos elementos acima da diagonal principal equivale a {somaElementos}')
```
