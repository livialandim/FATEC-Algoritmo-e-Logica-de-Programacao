### 4. Faça um programa que:
* receba o salário de um funcionário
* calcule e mostre o novo salário, sabendo-se que este sofreu um aumento de 25%.

      LEIA salario
      novoSalario = salario + (salario * 25 / 100)
      ESCREVA novoSalario

```
# O 'float' permite a entrada de todos os números reais
salarioInicial = float(input('Informe seu salário atual '))
# O '*' indica multiplicação
novoSalario = salarioInicial + ( salarioInicial * 25/100 )

print(f'Seu novo salário, com o aumento de 25% é {novoSalario:.2f}')
```

### 5. Faça um programa que receba o salário de um funcionário e o percentual de aumento, calcule e mostre o valor do aumento e o novo salário.


```
salarioInicial = float(input('Informe seu salário atual '))
aumentoPercentual = float(input('Informe o aumento percentual de seu salário '))
# A porcentagem não pode ser escrita com %, precisa ser /100, já que a % representa o resto da divisão
aumentoSalarial = salarioInicial * aumentoPercentual/100

novoSalario = salarioInicial + aumentoSalarial

print(f'Você teve um aumento salarial de {aumentoSalarial:.2f} e seu novo salário é {novoSalario:.2f}')
```

### 20. Faça um programa que receba um número real, encontre e mostre:

* a parte inteira desse número;
* a parte fracionária desse número;
* o arredondamento desse número.

      LEIA numero
      parte_inteira = numero // 1
      parte_fracionaria = numero - parte_inteira
      numero_arredondado = round (numero)
      ESCREVA parte_inteira
      ESCREVA parte_fracionaria
      ESCREVA numero_arredondado
      
**Observação**: Para arredondar um número em Python, usa-se a função **round(numero)**, onde se o número real/float estiver em igual distância entre o inteiro de cima e o inteiro de baixo, esta função arredonda para o número mais próximo.



```
numero = float(input('Informe um número real '))

parteInteira = numero // 1 
parteFracionaria = numero - parteInteira
# O round(...) é utilizado para arredondar um número
numeroArredondado = round (numero)

print('A parte inteira é ', parteInteira, ', a parte fracionária é ', parteFracionaria, ' e o número arredondado equivale a', numeroArredondado)
```

### 21. Faça um programa que receba uma hora formada por hora e minutos (um número real), calcule e mostre a hora digitada apenas em minutos. Lembre-se de que:
* para quatro e meia, deve-se digitar 4.30;
* os minutos vão de 0 a 59.
      LEIA hora
      h = pegar a parte inteira da variável hora
      minutos = hora − h
      conversao = (h * 60) + (minutos * 100)
      ESCREVA conversao


```
hora = float(input('Informe a hora e os minutos, onde 4h e 30min será escrito: 4.30 | '))
# O '//' serve para apresentar a divisão inteira
# A '%' apresenta o resto da divisão
h = hora // 1
minutos = hora - h
minutosInteiros = minutos * 100
# Os minutos serão multiplicados por 100 por causa que são a parte decimal do número fornecido
conversao = (h * 60) + minutosInteiros

# O f' .. {.. :,.2f} é utilizado para formatar as casas após a vírgula
print('A conversão de ', h, f' horas e {minutosInteiros:,.2f} minutos em apenas minutos é ', conversao)
```

### 22. Faça um programa que receba o custo de um espetáculo teatral e o preço do convite desse espetáculo. Esse programa deverá calcular e mostrar a quantidade de convites que devem ser vendidos para que, pelo menos, o custo do espetáculo seja alcançado.


```
custoEspetaculo = float(input('Informe o custo do espetáculo teatral '))
precoConvite = float(input('Informe o preço do convite desse espetáculo '))

import math 

quantidadeConvites = math.ceil(custoEspetaculo / precoConvite)

# math.ceil() arredonda o número para cima

print(f'A quantidade mínima de convites a ser vendida é {quantidadeConvites:.0f}')
```
