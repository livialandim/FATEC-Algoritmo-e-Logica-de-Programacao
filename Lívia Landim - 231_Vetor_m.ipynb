### 1. Faça um programa que calcule e apresente a média de idades de uma sala de 35 alunos.​​

```
i = 0
idade = []
somaIdade = 0
while i < 5 :
    idade.append(int(input('Informe a idade: ')))
    somaIdade = somaIdade + idade[i]
    i = i + 1
mediaIdade = somaIdade / i

print(f'A média de idades é {mediaIdade:.2f}')
```

### 9. Criar um algoritmo que leia dados para um vetor de 100 elementos inteiros, imprimir o maior, o menor, o percentual de números pares e a média dos elementos do vetor. Obs.: percentual = quantidade contada * 100 / quantidade total. Não use nenhuma função pronta da linguagem Python, a não ser len() e append().


```
numero = []
somaNumeros = 0
somaPares = 0
for i in range(5):
    numero.append(int(input('Informe um número inteiro: ')))

for i in range(len(numero)):
    if i == 0:
        maior = numero[i]
        menor = numero[i]
    else:
        if numero[i] >= maior:
            maior = numero[i]
        elif numero[i] <= menor:
            menor = numero[i]
    
    verificacaoPares = numero[i] % 2
    somaNumeros = somaNumeros + numero[i]

    if verificacaoPares == 0:
        somaPares = somaPares + 1

percentualPares = (somaPares * 100) / len(numero)
mediaNumeros = somaNumeros / len(numero)

print(f'O maior número é {maior} e o menor é {menor}, o percentual de números pares é {percentualPares} e a média de todos os números é {mediaNumeros}')
```
