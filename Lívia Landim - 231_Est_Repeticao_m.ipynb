###1. Faça um algoritmo  que mostre 10 vezes a frase “Bem vindo a Fatec!”. Pode implementar com o comando while ou for.

```
print('Exemplo do while.............\n\n')
# 1º passo - inicialização
contador = 1
# 2º passo - condição/critério de parada
while contador <= 5:
    print(contador, 'Bem vindo a Fatec!')
    contador = contador + 1 # 3º passo - incremento
```

```
print('\n\nExemplo do for.............\n')
# 1º passo - inicialização, 2º passo - condição/critério de parada
# e 3º passo - incremento, numa única linha no for
for contador in range(1,5):
    print(contador,'Bem vindo a Fatec!')
```

###6. Faça um algoritmo que receba a idade de 10 pessoas, calcule e exiba a quantidade de pessoas maiores de idade, sendo que a maioridade é obtida após completar 18 anos. Pode implementar com o comando while ou for.


```
print('----- Utilizando WHILE -----')
quantidadePessoas = 1
maioresIdade = 0
while quantidadePessoas < 11 :
    idade = int(input('Informe a idade '))
    if idade >= 18 :
        maioresIdade = maioresIdade + 1
        quantidadePessoas = quantidadePessoas + 1
    else :
        quantidadePessoas = quantidadePessoas + 1

print('No total são ', maioresIdade, ' maiores de idade')

```

```
print('----- Utilizando FOR -----')

maioresIdade = 0
for quantidadePessoas in range(1,11) :
    idade = int(input('Informe a idade '))
    if idade >= 18 :
        maioresIdade = maioresIdade + 1
        quantidadePessoas = quantidadePessoas + 1
    else :
        quantidadePessoas = quantidadePessoas + 1

print('No total são ', maioresIdade, ' maiores de idade')
```

###7. Escreva um algoritmo que receba 23 números, calcule e exiba a quantidade de números pares e impares. Pode implementar com o comando while ou for.


```
print('----- Utilizando WHILE -----')
quantidadeNumeros = 1
quantidadePares = 0
quantidadeImpares = 0

while quantidadeNumeros < 24 :
    numeros = int(input('Infome um número '))
    if numeros % 2 == 0 :
        quantidadePares = quantidadePares + 1
    else :
        quantidadeImpares = quantidadeImpares + 1
    quantidadeNumeros = quantidadeNumeros + 1
print('No total, são ', quantidadePares, ' números pares e ', quantidadeImpares, ' números ímpares')
```

```
print('----- Utilizando FOR -----')

quantidadeImpares = 0
quantidadePares = 0

for quantidadeNumeros in range(1,24) :
    numeros = int(input('Infome um número '))
    if numeros % 2 == 0 :
        quantidadePares = quantidadePares + 1
    else :
        quantidadeImpares = quantidadeImpares + 1
    quantidadeNumeros = quantidadeNumeros + 1
print('No total, são ', quantidadePares, ' números pares e ', quantidadeImpares, ' números ímpares')
```

###12. Faça um programa que receba a idade e a altura de 20 pessoas. Calcule e exiba a média das alturas das pessoas com mais de 20 anos.




```
 print('----- Utilizando WHILE -----')

pessoas = 0
somaAltura = 0
pessoasMaisVinte = 0
while pessoas < 3 :
    idade = int(input('Informe a idade '))
    altura = float(input('Informe a altura '))
    if idade > 20 :
        somaAltura = somaAltura + altura
        pessoasMaisVinte = pessoasMaisVinte + 1
        pessoas = pessoas + 1
    else: 
        pessoas = pessoas + 1
print(pessoasMaisVinte)
mediaAlturas = somaAltura / pessoasMaisVinte
print(mediaAlturas)
```

```
print('----- Utilizando FOR -----')

somaAltura = 0
pessoasMaisVinte = 0
for pessoas in range(3) :
    idade = int(input('Informe a idade '))
    altura = float(input('Informe a altura '))
    if idade > 20 :
        somaAltura = somaAltura + altura
        pessoasMaisVinte = pessoasMaisVinte + 1
        pessoas = pessoas + 1
    else: 
        pessoas = pessoas + 1
mediaAlturas = somaAltura / pessoasMaisVinte
print(f'A média das alturas das pessoas com mais {mediaAlturas:.2f}')
```

###17. Faça um programa que calcule e apresente as tabuadas do 3 ao 7. Este exercício utiliza DUAS estruturas de repetição.

```
print('----- Utilizando WHILE -----')

quantidadeNumeros = 3

while quantidadeNumeros <= 7 :
    valor = 0
    while valor < 11 :
        multiplicacao = valor * quantidadeNumeros
        print(multiplicacao)
        
        valor = valor + 1
    quantidadeNumeros = quantidadeNumeros + 1
    print('-'*50)
```

```
print('----- Utilizando FOR -----')

for quantidadeNumeros in range(3,8) :
    valor = 0
    for valor in range(11) :
        multiplicacao = valor * quantidadeNumeros
        print(multiplicacao)
        
        valor = valor + 1
    quantidadeNumeros = quantidadeNumeros + 1
    print('-'*50)
```

###18. Faça um programa para o curso de ADS (6 módulos), cada sala tem 30 alunos, calcule e apresente os seguintes itens:
*  Quantidade de homens e mulheres de cada módulo;
*  Média de idades de cada módulo;
*  Quantidade de homens e mulheres do curso todo;
*  Média de idades do curso todo.

Observação: este exercício utiliza DUAS estruturas de repetição.

```
somaIdadesCurso = 0
cursoContadorMasculino = 0
cursoContadorFeminino = 0 
for modulo in range(1,7) :
    print(modulo,'º módulo','.'*50)
    contadorFeminino = 0
    contadorMasculino = 0 
    somaIdades = 0
    for alunos in range(1,31):
        idade = int(input('Informe a idade '))     
        genero = input('Cadastre o gênero M/F: ')
        if genero == 'F' or genero == 'f':
            contadorFeminino = contadorFeminino + 1
            cursoContadorFeminino = cursoContadorFeminino + 1
        elif genero == 'M' or genero == 'm' :
            contadorMasculino = contadorMasculino + 1
            cursoContadorMasculino = cursoContadorMasculino + 1
        else :
            print('Gênero inválido!')

        somaIdades = somaIdades + idade
        mediaIdadeModulo = somaIdades / 30

        somaIdadesCurso = somaIdadesCurso + idade
        mediaIdadeCurso = somaIdadesCurso / 180
        
        alunos = alunos + 1
    print('Há ', contadorFeminino, ' mulheres e ', contadorMasculino, 'homens. A média de idades do ',modulo, 'º módulo é ', mediaIdadeModulo)
    modulo = modulo + 1
print('.'*50)

print('Quantidade de mulheres no curso ', cursoContadorFeminino, ' ; quantidade de homens no curso ', cursoContadorMasculino)
print('A média das idades do curso é ', mediaIdadeCurso)
```
